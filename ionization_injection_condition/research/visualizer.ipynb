##### Import libs

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import title, show
```

##### Input data

```python
#ctau = [30,40,50,60]
#ionization_degree = [12.12,15.82,19.38,22.77]
#trapping = [7.09,13.96,18.92,22.60]

ion_level = [6,7,8,9]
ionization_degree = [100,100,17.78,0.86]
trapping = [0,0,17.12,0.78]
```

##### Plot data

```python
fig, ax = plt.subplots(figsize=(20,10))
```

```python
plt.xlabel("Ionization niveau", fontdict={'fontsize':20})
plt.ylabel("Degree [%]", fontdict={'fontsize':20})
ax.plot(ion_level,trapping,label="Trapping degree", color="blue")
ax.plot(ion_level,ionization_degree,label="Ionization degree", color="green")

ax.grid(which='major', alpha=0.8, color='grey', linestyle='--', linewidth=0.6, animated='True')
ax.grid(which='minor', alpha=0.6, color='grey', linestyle='--', linewidth=0.5, animated='True')

for i in range(0,len(ion_level)):
    ax.scatter([ion_level[i]], [ionization_degree[i]], s=50, marker='x', color ='green')
    ax.annotate(ionization_degree[i], xy=(ion_level[i], ionization_degree[i]), xycoords='data', xytext=(+10, +10), textcoords='offset points', fontsize=16,arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2"))
    ax.scatter([ion_level[i]], [trapping[i]], s=50, marker='x', color ='blue')
    ax.annotate(trapping[i], xy=(ion_level[i], trapping[i]), xycoords='data', xytext=(+10, -25), textcoords='offset points', fontsize=16,arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2"))
legend = ax.legend(loc='upper left', shadow=True)
for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)  # the legend line width
title("Ionization degree/Trapping degree vs. argon ionization level  (laser duration: 40fs, a0 = 1.2 )", fontdict={'fontsize':20})
```

```python
plt.show()
```
